## task2

```python
weather['mean_prcp'] = weather.query('datatype == "PRCP"').groupby(level=0).mean()
```

```python
dp = weather.query('datatype=="PRCP"')
```

```python
dp
```

```python
value_binned = pd.cut(dp.mean_prcp, bins=3, labels=['low', 'med', 'high'])
value_binned.value_counts()
```

```python
dp[value_binned == 'high'].sort_values('mean_prcp', ascending=False)
```
